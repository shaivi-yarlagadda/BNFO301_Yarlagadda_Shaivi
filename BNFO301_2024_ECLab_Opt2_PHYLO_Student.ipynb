Problem 1. Create a definition that will randomly pick vectors, given a user defined number of clusters, K, and the sample data.  Return K random points contained within data

```python
#random starting points
def initialize_centroids(data, k):
    centroids_idx = np.random.choice(len(data), k, replace=False)
    centroids = data[centroids_idx]
    return centroids
```

Problem 2. Create a defintion that will calculate the euclidean distance between each centroid and each data point and assign each data point to the nearest cluster.  Return a list, where the outer list index is equivical to number of clusters K, inner list is assignment of each vector from data to its respective cluster

```python
#assign each vector to a cluster based on Euclidean distances
def assign_to_clusters(data, centroids):
    clusters = [[] for _ in range(len(centroids))]
    for point in data:
        distances = [np.linalg.norm(point - centroid) for centroid in centroids]
        nearest_centroid_idx = np.argmin(distances)
        clusters[nearest_centroid_idx].append(point)
    return clusters
```

Problem 3. Create a definition to find the average of each constituent vector assigned to a cluster and return mean for each cluster.  return cluster means for each cluster K

```python
#compute each cluster as average of its constituents
def update_centroids(data, clusters):
    centroids = []
    for cluster in clusters:
        if len(cluster) > 0:
            centroid = np.mean(cluster, axis=0)
            centroids.append(centroid)
        else:
            # If a cluster has no points assigned, randomly assign a new centroid
            centroids.append(data[np.random.randint(0, len(data))])
    return centroids
```

Problem 4. Create a definition to cluster each data point until means do not change, i.e. the difference is equal to 0. Return the clusters and nested list with assignment of each vector to each cluster K

```python
#iterate clusering
def kmeans(data, k):
    centroids = initialize_centroids(data, k)
    prev_centroids = np.copy(centroids)
    while True:
        clusters = assign_to_clusters(data, centroids)
        centroids = update_centroids(data, clusters)
        # Check for convergence
        if np.allclose(centroids, prev_centroids):
            break
        prev_centroids = np.copy(centroids)
    return clusters, centroids
```
